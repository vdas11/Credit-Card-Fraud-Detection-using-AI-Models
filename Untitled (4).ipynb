```python
## 1. Import Required Libraries
print("\n[1] Importing required libraries...")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
import warnings


warnings.filterwarnings('ignore')
print("Libraries imported successfully.")

## 2. Load Dataset
print("\n[2] Loading dataset...")
df = pd.read_csv('credit_card_fraud_optimized_dataset.csv')
print(f"Dataset loaded. Total rows: {df.shape[0]}, Total columns: {df.shape[1]}")
```

```python
## 3. Encode Categorical Features
print("\n[3] Encoding categorical features...")
categorical_cols = ['location', 'merchant', 'card_type', 'device_type']
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
print("Categorical features encoded.")
```

```python
## 4. Feature-Target Split
print("\n[4] Splitting features and target...")
X = df.drop(['transaction_id', 'user_id', 'is_fraud'], axis=1)
y = df['is_fraud']
print(f"Features shape: {X.shape}, Target shape: {y.shape}")
```

```python
## 5. Train-Test Split
print("\n[5] Performing train-test split...")

# Check and drop rows where target is NaN
print("\nChecking for missing values:")
print(df.isna().sum())

df = df.dropna(subset=['is_fraud'])

X = df.drop(['transaction_id', 'user_id', 'is_fraud'], axis=1)
y = df['is_fraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print(f"Training set: {X_train.shape}, Test set: {X_test.shape}")
```

```python
## 6. Scale Numerical Features
print("\n[6] Scaling numerical features...")
scaler = StandardScaler()
X_train[['time', 'amount', 'previous_fraud']] = scaler.fit_transform(X_train[['time', 'amount', 'previous_fraud']])
X_test[['time', 'amount', 'previous_fraud']] = scaler.transform(X_test[['time', 'amount', 'previous_fraud']])
print("Numerical features scaled.")
```

```python
## 7. Define Models
print("\n[7] Defining models...")
models = {
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42),
    'CatBoost': CatBoostClassifier(verbose=0, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'LightGBM': LGBMClassifier(random_state=42)
}
print(f"Defined {len(models)} models: {list(models.keys())}")
```

```python
## 8. Train and Evaluate Models
print("\n[8] Training and evaluating models...")
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    preds = model.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, preds)
    results[name] = auc
    print(f"{name} AUC: {auc:.4f}")
```

```python
print("Actual frauds in entire dataset:", int(df['is_fraud'].sum()))

for name, model in {
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'LightGBM': LGBMClassifier(random_state=42)
}.items():
    model.fit(X, y)
    preds = model.predict(X)
    print(f"{name} predicted frauds: {int(np.sum(preds))}")
```

```python
## 9. Select and Evaluate Best Model
print("\n[9] Selecting best model based on AUC...")
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"Best model: {best_model_name} with AUC = {results[best_model_name]:.4f}")

y_pred = best_model.predict(X_test)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

## 11. Plot ROC Curve for Best Model
from sklearn.metrics import roc_curve, auc

print("\n[11] Plotting ROC curve...")

# Predict probabilities
y_probs = best_model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'ROC Curve - {best_model_name}')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
## 10. Plot Feature Importances
print("\n[10] Plotting feature importances...")
if hasattr(best_model, 'feature_importances_'):
    importance = pd.Series(best_model.feature_importances_, index=X.columns)
    importance.nlargest(10).plot(kind='barh')
    plt.title(f'{best_model_name} Feature Importance')
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.tight_layout()
    plt.show()
else:
    print("Feature importances not available for this model.")
```

```python
## 12. Fraud Predictions vs Actual - Bar Chart
print("\n[12] Plotting fraud prediction counts for each model...")

models_for_plot = {
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    'CatBoost': CatBoostClassifier(verbose=0, random_state=42),
    'LightGBM': LGBMClassifier(random_state=42)
}

fraud_counts = {}
actual_fraud = y_test.sum()

for name, model in models_for_plot.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    predicted_fraud = np.sum(preds)
    fraud_counts[name] = predicted_fraud

# Add actual fraud to the plot
fraud_counts['Actual Fraud'] = int(actual_fraud)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(fraud_counts.keys(), fraud_counts.values(), color=['skyblue', 'orange', 'limegreen', 'violet', 'gold', 'red'])
plt.title("Fraud Predictions vs Actual Fraud Count")
plt.ylabel("Number of Fraudulent Transactions")
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
